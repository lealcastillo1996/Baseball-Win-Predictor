# Loading dataset

```python
#Importing pandas, numpy
import pandas as pd
import numpy as np
#Importing pathlib
import pathlib
pathlib.Path.cwd()
#display all columns
pd.pandas.set_option ("display.max_columns", None)

#reading csv
path = pathlib.Path.cwd() / 'sets' / 'mlb_dataset.csv'
df = pd.read_csv(path, index_col = None)

#Initial describe
print(df.describe())

df
```

# Missing values


```python
dataset = df.copy()
#Checking the % of NAN in file
features_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]

numerical_features_na = [features for features in features_with_na if dataset[features].dtypes != "O"]
categorical_features_na = [features for features in features_with_na if dataset[features].dtypes == "O"]


#Print missin values numerical val
print("Numerical Missing Features %")
for feature in numerical_features_na:
    print((feature,np.round(dataset[feature].isnull().mean(),4)*100), "% MISSING VALUES")
    
    
print("Categorical Missing Features %")
for feature in categorical_features_na:
    print((feature,np.round(dataset[feature].isnull().mean(),4)*100), "% MISSING VALUES")
```

Since missing values are only present in the feature we are looking for, we will not do any missing value handle and just will drop NaN

```python
df = df.dropna()
df= df.reset_index(drop=True)
df
```

# Temporal relationship

```python
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
dft = df.copy()
#Date to date format
dft['Fecha'] =  pd.to_datetime(dft['Fecha'], format='%Y%m%d')
dft

#Creating a season (year column)
dft['Season'] =  pd.DatetimeIndex(dft['Fecha']).year
dft

#Creating a pivot table to summarize each year local win %
pivot = pd.pivot_table(dft, index = ["Season"], values= "Local Win?", aggfunc=["mean"], fill_value= 0, margins=True, margins_name = "Sum Total" )
pivot

p1 = pivot.iloc[:-1,0:1]*100
p1
plt.style.use('seaborn-dark')
a =p1.plot()
plt.xlabel("Year")
plt.ylabel("% of games won by locals")
plt.title("% of games won by locals vs Season")
a
```

We can observe that % of local win has varied between 54.5% and 52% over the analyzed years

# Correlation Matrix

```python
#Defining correlation function 
def cor(x_train,val):
    
    #Creating correlation matrix
    corr_matrix = x_train.corr()
    
    sort = ((corr_matrix["Local Win?"].sort_values(ascending=False)))
    sort = abs(sort)
    
    
    
    
    #Sorting the list
    sort = sort.sort_values(ascending=False)
    sort
    #Priinting entire list
    print(sort)

    sort= sort.iloc[:].index.values
    selected = list(sort)
    
    
    
    #Printing correlation matrix heatmap
    
    # import modules

    sns.set_style("whitegrid")
    #Selecting top variables in a list
    selected = selected[:val]
    #Selecting top variables in original correlation matrix
    cor_matrix_sel = corr_matrix.loc[selected,selected]
    
   
    

    # plotting correlation heatmap for top 10 features
    print("Correlation Matrix Heatmap")
    dataplot = sns.heatmap(cor_matrix_sel, cmap="viridis", annot=False)

    # displaying heatmap
    plt.show()
    return selected

cor_list =cor(dft,10)
print("Correlation Matrix method")
```
