# Correlation Matrix

```python
#Defining correlation function 
def cor(x_train,val):
    
    #Creating correlation matrix
    corr_matrix = x_train.corr()
    
    sort = ((corr_matrix["Local Win?"].sort_values(ascending=False)))
    sort = abs(sort)
    
    
    
    
    #Sorting the list
    sort = sort.sort_values(ascending=False)
    sort
    #Priinting entire list
    print(sort)

    sort= sort.iloc[:].index.values
    selected = list(sort)
    
    
    
    #Printing correlation matrix heatmap
    
    # import modules

    sns.set_style("whitegrid")
    #Selecting top variables in a list
    selected = selected[:val]
    #Selecting top variables in original correlation matrix
    cor_matrix_sel = corr_matrix.loc[selected,selected]
    
   
    

    # plotting correlation heatmap for top 10 features
    print("Correlation Matrix Heatmap")
    dataplot = sns.heatmap(cor_matrix_sel, cmap="viridis", annot=False)

    # displaying heatmap
    plt.show()
    return selected

cor_list =cor(dft,10)
print("Correlation Matrix method")
```

# Conclusion number 1
Features recollected does not correlate very well with the final result, we will try to create new features with the scores concept introduced in ....

New score features will be proposed to try to find a correlation


```python
#Cleaning possible NAN strings
for feature in con_features:
    
   
        
    try:
        df= df[df[feature].str.contains("NaN")!=True]

    except Exception as e:
        print("not containing nan",e) 
```

```python
#Total score performance
df=dft.copy()
#Total scored runs of local team - total scored runs of visit team
df["loc score"] = df["Local Last 5 scored at home"] -  df["Visit Last 5 scored at visit"]


#Bullpen perfomance

df["loc bull"] = (df["Local Bullpen recieved last 5 games"] / df["Local Bullpen in"]) * (1-df["Local pitcher in avg"])
df["vis bull"] = (df["Visit Bullpen recieved last 5 games"]  / df["Visit Bullpen in"]) * (1-df["Visit pitcher in avg"])
#Local bullpen -  Visit bullpen performance
df["bull"] = (df["loc bull"]) - (df["vis bull"])


#Pitcher Performance
df["Local pitcher era"] = ((df["Local pitcher recieved last 5 games" ]) / df["Local pitcher in"]) 
df["Visit pitcher era"] = ((df["Visit pitcher recieved last 5 games" ]) / df["Visit pitcher in"]) 
#Pitcher kk
df["loc pitcher kk"] = ((df["Local pitcher k"] / df["Local pitcher in"])) 
df["vis pitcher kk"] = (df["Visit pitcher k"] / df["Visit pitcher in"])
#Pitcher score
df["pitch"] = (df["loc pitcher kk"] - df["Local pitcher era"]) - (df["vis pitcher kk"] - df["Visit pitcher era"])



#Last 5 results factor
df["factor loc"] = (df["Local Last 5 form at home"] - df["Visit Last 5 form at visit"])

#Last 10 results factor
df["factor loc 10"] = (df["Local last 10 form"] - df["Visit last 10 form"])


columnsTitles = ["Fecha","Local","Visit","Local pitcher","Visit pitcher","pitch","loc bull","vis bull","factor loc","factor loc 10", "Direct local win?" , "Local Win?"]


df = df.reindex(columns=columnsTitles)


df.dropna(inplace=True)
df.reset_index(inplace=True,drop=True)


#Saving resulting df to a csv
path = pathlib.Path.cwd() / 'sets' / 'mlb_dataset_treated2.csv'
df.to_csv(path, index= False)



df
```
