#  Naive-Bayes Classificator

Naive- Bayes classicator algorithm was choosed because is among one of the best unflexible classificators when large variance is observed in data.

```python
#Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Importing pathlib
import pathlib
pathlib.Path.cwd()
#display all columns
pd.pandas.set_option ("display.max_columns", None)
# to make this notebook's output stable across runs
np.random.seed(42)
from sklearn.naive_bayes import GaussianNB
```

```python
#Csv import

#Train x
path = pathlib.Path.cwd() / 'sets' / 'xtrain_ready.csv'
X_train = pd.read_csv(path, index_col = None)

#Train y
path = pathlib.Path.cwd() / 'sets' / 'ytrain_ready.csv'
y_train = pd.read_csv(path, index_col = None)

#Test x
path = pathlib.Path.cwd() / 'sets' / 'xtest_ready.csv'
X_test = pd.read_csv(path, index_col = None)


#Test y
path = pathlib.Path.cwd() / 'sets' / 'ytest_ready.csv'
y_test = pd.read_csv(path, index_col = None)



#Test team names
path = pathlib.Path.cwd() / 'sets' / 'xtest_teams.csv'
test_teams = pd.read_csv(path, index_col = None)

#Checking if training data contains NaN
X_train.isnull().any()











```

# Preparing output

```python
#DF to numpy arrays
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
X_test = X_test.to_numpy()
y_test = y_test.to_numpy()



#Output True if  == 1 else False
y_train_T = (y_train == 1)
y_test_T = (y_test == 1)



# Displaying number of True and False in the train set
pd.DataFrame(y_train_T).value_counts()
```

# Model Training

```python
#Importing Naive-Bayes classifier 
from sklearn.naive_bayes import GaussianNB
#Importing grid hyperparameter search
from sklearn.model_selection import GridSearchCV
#Logspace utility
from sklearn.utils.fixes import loguniform



#Defining hyperparameters grid
param_grid_nb = {
    'var_smoothing': np.logspace(0,-9, num=100)
}

#model grid search
nbModel_grid = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid_nb, verbose=1, cv=10, n_jobs=-1)


#Fitting model to train data
nbModel_grid.fit(X_train, y_train_T)


#Defining the best model
nbModel = nbModel_grid.best_estimator_
print(nbModel)

```

# Model metrics and prediction in test set

```python
from sklearn.metrics import f1_score
#Checking model accuracy
print (f'Train Accuracy - : {nbModel.score(X_train,y_train_T):.3f}')
print (f'Test Accuracy - : {nbModel.score(X_test,y_test_T):.3f}')



#Prediction in test set
test_prediction = nbModel.predict(X_test)

#Scores of positive case
scores= nbModel.predict_proba(X_test)[:,1]

```

```python


#F1 SCORE IN TEST SET
print("F1 score in test set",round(f1_score(y_test_T,test_prediction),3))

#Confusion Matrix of prediction
from sklearn.metrics import confusion_matrix
pred= test_prediction.copy()
exp = y_test_T
con_mat= confusion_matrix(exp,pred)
print(con_mat)
print (con_mat[1,1]," Correct guess out of", (con_mat[1,1] + con_mat[0,1]), "Predictions for Local team win", "Acurracy of: ", 
       round(con_mat[1,1]/((con_mat[1,1] + con_mat[0,1])),3), "%" )
print (con_mat[0,0]," Correct guess out of", (con_mat[1,0] + con_mat[0,0]), "Predictions for Visit team win","Acurracy of: ", 
       round(con_mat[0,0]/((con_mat[1,0] + con_mat[0,0])),3), "%" )

#Plotting confusion matrix
plt.style.use('seaborn-dark')
ax= plt.subplot()
sns.heatmap(con_mat, annot=True, fmt='g', ax=ax,  cmap= "viridis");  #annot=True to annotate cells, ftm='g' to disable scientific notation

# labels, title and ticks

ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix for test set'); 
ax.xaxis.set_ticklabels(['Visit Win', 'Local Win']); ax.yaxis.set_ticklabels(['Visit Win', 'Local Win']);


#df with predictions
df = pd.DataFrame()
df = pd.DataFrame(test_teams.copy())
df["P_score"] = scores
df["Local win Predicted"] = pred
df["Local win Exp"] = exp
df.head(30)

```

# Adjusting Threshold 

Adjusting original threshold (0.5 by default) to increase accuracy in an specific result can be done, of course a sacrifice in recall will be observed but is not a matter of concern since we are looking for the highest precission posible for this classification.

```python
#Looking for the  histogram of predicted Yscores

sns.displot(df["P_score"])




```

```python
#Looking for mean y score predicted vs true outcome
ax = sns.boxplot(x="Local win Exp", y="P_score" , data= df)
#Limits
ax.set(ylim=(.4, .7))
```

We can observe there are clear thresholds for expected false (+-.48) and expected true (+- .61), so we will use this thresholds to predict better when 

# Local Win prediction with a better Threshold

```python

#Changing threshold affects the decision
threshold = .61
y_threshold_pred = (scores > threshold)
y_threshold_pred

#df with predictions
df = pd.DataFrame()
df = pd.DataFrame(test_teams.copy())
df["P_score"] = scores
df["Local win Predicted"] = y_threshold_pred
df["Local win Exp"] = exp
#Just keeping true predictions
df = df.loc[df["Local win Predicted"] == True]
df.head(30)

con_mat= confusion_matrix(df["Local win Exp"],df["Local win Predicted"])
print(con_mat)
print (con_mat[1,1]," Correct guess out of", (con_mat[1,1] + con_mat[0,1]), "Predictions for Local team win", "Acurracy of: ", 
       round(con_mat[1,1]/((con_mat[1,1] + con_mat[0,1])),3), "%" )
df
```

We can observe an important boost of acurracy when introducing the threshold

# Visit Win prediction with a better Threshold

```python
#Changing threshold affects the decision
threshold = .48
y_threshold_pred = (scores > threshold)
y_threshold_pred

#df with predictions
df = pd.DataFrame()
df = pd.DataFrame(test_teams.copy())
df["P_score"] = scores
df["Local win Predicted"] = y_threshold_pred
df["Local win Exp"] = exp
#Just keeping true predictions
df = df.loc[df["Local win Predicted"] == False]
df.head(30)

con_mat= confusion_matrix(df["Local win Exp"],df["Local win Predicted"])
print(con_mat)
print (con_mat[0,0]," Correct guess out of", (con_mat[1,0] + con_mat[0,0]), "Predictions for Visit team win","Acurracy of: ", 
       round(con_mat[0,0]/((con_mat[1,0] + con_mat[0,0])),3), "%" )

df
```

A small boost in accuracy was observed, however threshold can be changed to adjust more the results sacrifying even more the recall.
